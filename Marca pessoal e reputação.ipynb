#  Marca e reputação é a mesma coisa?

```python
import matplotlib.pyplot as plt
labels = 'Marca Pessoal é como você se comporta', 'Reputação é como as pessoas te julgam', 'Um complementa o outro', 'A marca é como outros te enxergam','Marca pessoal é um esforço'             
sizes = [20000.000,30000.000, 20000.000, 10000.000, 20000.000 ]
fig1, ax1 = plt.subplots(figsize=(6, 6))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
